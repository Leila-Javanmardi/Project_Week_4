### Research Question 2- How does food waste vary across European countries, and what are the primary sources of food waste (household, retail, food service)? 


```python
fil_FW.columns
```

```python

# Plotting food waste across European countries
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='combined figures (kg/capita/year)', data=fil_FW)
plt.xticks(rotation=90)
plt.title('Food Waste across European Countries')
plt.xlabel('Country')
plt.ylabel('Food Waste (kg/capita/year)')
plt.show()
```

```python
# Plotting primary sources of food waste
sources_columns = ['Household estimate (kg/capita/year)', 'Retail estimate (kg/capita/year)',
                   'Food service estimate (kg/capita/year)']
sources_labels = ['Household', 'Retail', 'Food Service']

plt.figure(figsize=(10, 6))
sns.boxplot(data=fil_FW[sources_columns])
plt.title('Primary Sources of Food Waste')
plt.xlabel('Source')
plt.ylabel('Food Waste (kg/capita/year)')
plt.xticks(range(len(sources_columns)), sources_labels)
plt.show()
```
