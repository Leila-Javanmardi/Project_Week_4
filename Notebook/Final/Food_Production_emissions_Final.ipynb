```python
plt.figure(figsize=(12, 8))

sns.barplot(x='Country', y='Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)', 
            hue='Food product', 
            data=EU_food_production_Emissions)

sns.barplot(x='Country', y='Production (tonnes)', 
            hue='Food product', 
            data=EU_food_production_Emissions, 
            alpha=0.5)  # To make the bars semi-transparent

plt.title('Greenhouse Gas Emissions vs. Food Production by Country')
plt.xlabel('Country')
plt.ylabel('Greenhouse Gas Emissions per 1000kcal / Production (tonnes)')
plt.legend(title='Food product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# european regions
eastern_europe = ['Albania', 'Belarus', 'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Czechia', 'Hungary', 'Montenegro', 'North Macedonia', 'Poland', 'Romania', 'Serbia', 'Slovakia', 'Slovenia', 'Ukraine']
southern_europe = ['Greece', 'Italy', 'Portugal', 'Spain', 'Malta']
western_europe = ['Austria', 'Belgium', 'France', 'Germany', 'Ireland', 'Luxembourg', 'Netherlands', 'Switzerland']
northern_europe = ['Denmark', 'Estonia', 'Finland', 'Latvia', 'Lithuania', 'Sweden']

# separate plots for each region
regions = {'Eastern Europe': eastern_europe, 'Southern Europe': southern_europe, 'Western Europe': western_europe, 'Northern Europe': northern_europe}

for region, countries in regions.items():
    # Filter 
    region_df = EU_food_production_Emissions[EU_food_production_Emissions['Country'].isin(countries)]
    
    # Plotting greenhouse gas emissions vs. food production as a grouped bar chart
    plt.figure(figsize=(12, 8))

    sns.barplot(x='Country', y='Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)', 
                hue='Food product', 
                data=region_df)

    sns.barplot(x='Country', y='Production (tonnes)', 
                hue='Food product', 
                data=region_df, 
                alpha=0.5)  # To make the bars semi-transparent

    plt.title(f'Greenhouse Gas Emissions vs. Food Production in {region}')
    plt.xlabel('Country')
    plt.ylabel('Greenhouse Gas Emissions per 1000kcal / Production (tonnes)')
    plt.legend(title='Food product', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python

# Calculate emissions per tonne of production
EU_food_production_Emissions['Emissions_per_tonne'] = EU_food_production_Emissions['Total_emissions'] / EU_food_production_Emissions['Production (tonnes)']

# Get unique food products
food_products = EU_food_production_Emissions['Food product'].unique()

# Iterate over each food product and create separate graphs
for food_product in food_products:
    # Filter data for the current food product
    food_df = EU_food_production_Emissions[EU_food_production_Emissions['Food product'] == food_product]
    
    # Plotting emissions per tonne of production for each country
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Country', y='Emissions_per_tonne', data=food_df)
    plt.title(f'Emissions per Tonne of Production for {food_product}')
    plt.xlabel('Country')
    plt.ylabel('Emissions per Tonne of Production')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
